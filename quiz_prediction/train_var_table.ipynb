```python
master = pd.DataFrame()
for week_num in NEW_VAR_MAP:
    if week_num != "post":
        temp = pd.DataFrame([[week_num, item[0], val] for item in NEW_VAR_MAP[week_num]["Quant"].items() for val in item[1]])
        temp.columns = ["week", "construct", "var"]
        master = pd.concat([master, temp], axis=0)
display(master)
```

```python
temp = pd.concat([df_all['Pre & Post'][["Var", "Item", "Label", "Response Values"]], df_all['Weekly'][["Var", "Item", "Label", "Response Values"]]], axis=0)
master = master.merge(temp, left_on="var", right_on="Var", how="left")
master = master.drop("Var", axis=1)
```

```python
pre_construct = NEW_VAR_MAP['week_1']["Quant"].keys()
post_construct = NEW_VAR_MAP['post']["Quant"].keys()
df_pre = pd.DataFrame(['T' for i in range(len(pre_construct))], index = pre_construct, columns=["Pre"])
df_post = pd.DataFrame(['T' for i in range(len(post_construct))], index = post_construct, columns = ["Post"])
df_comp = pd.concat([df_pre, df_post], axis=1)
print("Num Common Construct:  {}".format(len(set(pre_construct).intersection(set(post_construct)))))
print("Pre Uniq Construct:  {}".format(list(df_comp[df_comp["Post"].isna()].index)))
print("Post Uniq Construct:  {}".format(list(df_comp[df_comp["Pre"].isna()].index)))
display(df_comp)
```
