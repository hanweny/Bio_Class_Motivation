```python
df = pd.read_csv("../data/class_data.csv")
df_proc = preprocess_df(df, verbose = False)
```

#### Create correlation matrix for each week

```python
def plot_corr(df, size = (10, 5)):
    corr = df.corr()
    f = plt.figure(figsize=size)
    plt.matshow(corr, fignum=f.number)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical')
    plt.yticks(range(len(corr.columns)), corr.columns)
    cb = plt.colorbar()
    plt.clim(-1, 1)
    cb.ax.tick_params(labelsize=12)
    plt.show()
    
for week_num in range(1, 6):
    print("Week", week_num)
    curr_var_map = NEW_VAR_MAP["week_{}".format(week_num)]["Quant"]
    curr_var_list = [j for i in curr_var_map.values() for j in i]
    plot_corr(df_proc[curr_var_list], (19, 15))
```

#### Create correlation matrix for each week by Construct

```python
def plot_corr_by_week_construct(week_num):
    week_num = "week_{}".format(week_num)
    print(week_num, '\n')
    for c in NEW_VAR_MAP[week_num]["Quant"]:
        print("Construct {}".format(c))
        print(list(NEW_VAR_MAP[week_num]["Quant"][c]))
        plot_corr(df_proc[list(NEW_VAR_MAP[week_num]["Quant"][c])])
```
